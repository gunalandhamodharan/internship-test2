```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('input/main.csv')
data.tail()
```

```python
#QUESTION 01 (Filling NA values)

data.groupby('product_description')['price'].mean()

data["price"] = data['price'].fillna(data.groupby('product_description')['price'].transform("mean"))
```

```python
data["sale_amount"] = data["sales_quantity"].where(
    data["unit"] != "pcs",
    data["product_description"].str.split("-", expand=True)[1].astype("float") * \
    data["sales_quantity"] 
)
```

```python
data.to_csv('main.csv')
```
